```python
# import libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Write a Pandas program to compare the elements of two Pandas series. 
nb = pd.Series([4, 65, 436, 3, 9])
na = pd.Series([7, 0, 3, 897, 9])
z = nb == na
print(z)
```

```python
# Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
m = pd.Series([2, 4, 6, 8, 14])
n = pd.Series([1, 3, 5, 7, 9])
addition = m + n
print("Addition:")
print(addition)
subtraction = m - n
print("\nSubtraction:")
print(subtraction)
multiplication = m * n
print("\nMultiplication:")
print(multiplication)
division = m / n
print("\nDivision:")
print(division)
```

```python
# Write a Pandas program to convert a dictionary to a Pandas series. 
dic = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}
p = pd.Series(dic)
print(p)
```

```python
# Write a Pandas program to convert a given series to an array.
q = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])
w = np.array(q)
print(w)
```

```python
# Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series.
df=pd.read_csv("AfricaCupofNationsMatches.csv")
y = df['HomeTeamGoals'].mode()[0]
df['HomeTeamGoals'] = df['HomeTeamGoals'].where(df['HomeTeamGoals'] == y, 'Other')
print(df['HomeTeamGoals'])
```
